```python

import pandas as pd
#baixar os csv desse link ;
#https://drive.google.com/drive/folders/11_GBWJRvaeNmtXbMjM182Mx72lNOtfKb
```

```python
df = pd.concat(
    map(pd.read_csv, ['0.csv', '1.csv','2.csv','3.csv','4.csv','5.csv','6.csv','7.csv','8.csv']), ignore_index=True)
print(df)
```

```python
df.info()
```

```python
#fazer um clone 
df_clone=df[:]
```

```python
#escolher as colunas que vao ser passadas para do df final e depois adicionar uma nova coluna com o engenho de busca
df_desejado=df_clone[['Title','Authors','Publication year','Abstract','URLs','Item type']]
```

```python
df_desejado=df_desejado.assign(SEARCH_ENGINE='ACM')
df_desejado.rename(columns={'Title': 'TITLE','Authors':'AUTHORS','Publication year':'PUBLICATION_YEAR','Abstract':'ABSTRACT','URLs':'LINKS','Item type':'ITEM TYPE'  })
```

```python
df_desejado.info()
```

```python
df_desejado.to_csv("ACM_Completo_Normalizado.csv", encoding = 'utf-8')
```
